```
import keras
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```
keras.backend.backend()
```

```
fm=keras.datasets.fashion_mnist
(X_train,y_train),(X_test,y_test)=fm.load_data()
```

```
X_train=X_train/255
```

```
X_test=X_test/255
```

```
from keras.models import Sequential
from keras.layers import Flatten,Dense,Activation
model=Sequential()
model.add(Flatten(input_shape=[28,28]))
```

```
model.summary()
```

```
model.add(Dense(100,activation="relu"))
```

```
model.summary()
```

```
model.add(Dense(10,activation="softmax"))
```

```
model.summary()
```

```
model.compile(loss="sparse_categorical_crossentropy",optimizer="adam",metrics="accuracy")
```

```
model.fit(X_train,y_train)
```

```
model.evaluate(X_test,y_test)
```

```
plt.matshow(X_test[2])
```

```
yp=model.predict(X_test)
```

```
np.argmax(yp[2])
```

```
dig=keras.datasets.mnist
(X_train,y_train),(X_test,y_test)=dig.load_data()
```

```
X_train=X_train/255
X_test=X_test/255
```

```
modeld=Sequential()
modeld.add(Flatten(input_shape=[28,28]))
modeld.add(Dense(20,activation="relu"))
modeld.add(Dense(10,activation="softmax"))
```

```
modeld.summary()
```

```
modeld.compile(loss="sparse_categorical_crossentropy",optimizer="adam",metrics=["accuracy"])
```

```
modeld.fit(X_train,y_train)
```

```
modeld.evaluate(X_test,y_test)
```
